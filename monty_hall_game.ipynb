```python
import numpy as np
import random
```

```python
def monty_hall(switch):
    doors = np.array([0,0,0])  #all the doors
    winner_index = np.random.randint(0,3)  #Random number between 0 and 3 exclusive 3
    doors[winner_index] = 1 #placing the car at the winner index in the doors array
    choice = random.randint(0,3)
    openable_doors = [i for i in range(3) if i not in (choice,winner_index)]  #we are not opening the door that consist of the prize or car by  ourselves
    host_open = [i for i in openable_doors][0]   #then the host opens the door that doesnot consist of the car
    if switch:
        choice = [i for i in range(3) if i not in (host_open,choice)]    #we can only open the door that is not opened by host and by us previously if we switch 
    return doors[choice]    
    
        
    
```
